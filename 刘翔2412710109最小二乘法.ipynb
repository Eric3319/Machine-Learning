```python
import matplotlib.pyplot as plt

# 数据点
x_data = [1, 2, 3, 4, 5]
y_data = [2, 4, 6, 8, 10]

n = len(x_data)

sum_x = 0
sum_y = 0
sum_xy = 0
sum_x_squared = 0

for i in range(n):
    sum_x += x_data[i]
    sum_y += y_data[i]
    sum_xy += x_data[i] * y_data[i]
    sum_x_squared += x_data[i]**2

a = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)
b = (sum_y - a * sum_x) / n

print(f"拟合的直线方程为 y = {a:.2f}x + {b:.2f}")

# 绘制数据点和拟合直线
plt.scatter(x_data, y_data, color='blue', label='Data Points')
fit_y = [a*x + b for x in x_data]
plt.plot(x_data, fit_y, color='red', label='Fitted Line')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
